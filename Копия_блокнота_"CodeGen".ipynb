```python
%pip install transformers==4.25.1
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
from transformers import AutoTokenizer, AutoModelForCausalLM
DEVICE = torch.device("cuda") if torch.cuda.is_available() else None
tokenizer = AutoTokenizer.from_pretrained("Salesforce/codegen-2B-mono")
model = AutoModelForCausalLM.from_pretrained("Salesforce/codegen-2B-mono").to(DEVICE)
```

```python
#@title Generate code

code = "def hello_world():" #@param {type:"string"}
len = 24 #@param {type:"integer"}

inputs = tokenizer(code, return_tensors="pt").to(0)
sample = model.generate(**inputs, max_length=len)
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```

```python
#@title Prod generater
from datetime import datetime

def generate(code, len=154):
  start = datetime.now()
  inputs = tokenizer(code, return_tensors="pt").to(DEVICE)
  print(f"Tokenize: {datetime.now() - start}")
  sample = model.generate(**inputs, max_length=len).to(DEVICE)
  print(f"Generate: {datetime.now() - start}")
  return tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"])

print(generate('extract only numbers from  DataFrame'))
```
